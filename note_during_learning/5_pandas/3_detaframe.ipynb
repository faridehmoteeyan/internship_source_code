###  Relational Operations

```python
dummy_data1 = {
        'id': ['1', '2', '3', '4', '5'],
        'Feature1': ['A', 'C', 'E', 'G', 'I'],
        'Feature2': ['B', 'D', 'F', 'H', 'J']}
df1 = pd.DataFrame(dummy_data1)
df1
```

```python
dummy_data2 = {
        'id': ['1', '2', '6', '7', '8'],
        'Feature1': ['K', 'M', 'O', 'Q', 'S'],
        'Feature2': ['L', 'N', 'P', 'R', 'T']}
df2 = pd.DataFrame(dummy_data2, columns = ['id', 'Feature1', 'Feature2'])
df2
# import: column must have same name as key in dicthionary
```

```python
dummy_data3 = {
        'id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'Feature3': [12, 13, 14, 15, 16, 17, 15, 12, 13, 23]}
df3 = pd.DataFrame(dummy_data3, columns = ['id', 'Feature3'])
df3
```

```python
df1
```

####  concatenate 

```python
conc_1 = pd.concat([df1,df2])
conc_1
```

```python
conc_2 = pd.concat([df1,df2],ignore_index=1)
conc_2
```

#### merge 

```python
df_merge_col = pd.merge(conc_1,df3,on='id')
df_merge_col.sort_values('Feature3', ascending=False)
```

```python
df_outer = pd.merge(df1, df2, on='id', how='outer')

df_outer
```

```python
df_inner = pd.merge(df1, df2, on='id', how='inner')

df_inner
```

```python
df_right = pd.merge(df1, df2, on='id', how='right')

df_right

```

```python
df_left = pd.merge(df1, df2, on='id', how='left')

df_left
```
