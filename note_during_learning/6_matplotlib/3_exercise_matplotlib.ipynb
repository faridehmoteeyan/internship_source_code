```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(100)
y=2*x + 1
fig = plt.figure()
plt.plot(x,y,'c-1')
plt.legend(["y=2*x + 1"],loc = "right")
plt.ylim(0,70)
plt.ylabel("kilos",fontsize = 90)
```

```python
import math

x = np.arange(-10, 10, 1)
y = 1 / (1 + (math.e)**-x)

plt.plot(x, y,'*')
plt.show()

```
