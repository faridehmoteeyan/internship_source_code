#### marge 12 csv to one csv 

```python
files = [file for file in os.listdir(str(os.getcwd())+"\\Sales_Data") if not file.startswith('.')]#ignore hidden files
files
all_months_data = pd.DataFrame()

for file in files:
    current_file = pd.read_csv("Sales_Data\\" + "Sales_April_2019.csv")
    all_months_data = pd.concat([all_months_data, current_file])

#os.mkdir("annual_sales")
all_months_data.to_csv("annual_sales\\annual_sales.csv")
```
